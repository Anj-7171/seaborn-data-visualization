# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
Seaborn divides plot into the below categories –

*   Relational plots
*   Categorical plots
*   Distribution plots
*   Matrix plots

1. **Relational plots**

Relational Plots are used for visualizing the statistical relationship between the data points.

```python
sns.get_dataset_names()
```

```python
import seaborn as sns

# reading the dataset
tips = sns.load_dataset('tips')

# printing first five entries
print(tips.head())
```

```python
#Understanding the data
print(tips.nunique())
print(tips.sex.unique())
```

```python
print(tips.day.unique())
```

To draw the relational plots seaborn provides three functions. These are:

scatterplot()

relplot()

lineplot()

**Case 1**

```python
# plotting a simple visualization between total bill and tips
sns.scatterplot(x ="total_bill", y ="tip",data=tips)
```

**Case 2**

```python
# Adding the sex column to understand the data better

sns.relplot(x="total_bill", y="tip", hue="day", data=tips)                      #hue : This parameter take column name for colour encoding.
```

**Case 3**

```python
# Similarly we will understand the relationship by time

sns.relplot(x="total_bill", y="tip", hue="time", data=tips)
```

**Case 4**

```python
# Split the chart based on colum Sex

sns.relplot(x="total_bill", y="tip", hue="time", col= "sex",data=tips)
```

2. **Categorical plots**

Here the variables can be either be completely numerical or a category. Main Categorical plots  are:

barplot()

countplot()

boxplot()



```python
# Bar Plots


sns.barplot(x ='sex', y ='total_bill', data = tips)
```

```python
# Countplot - A countplot  counts the categories and returns a count of their occurrences.

sns.countplot(x ='day', data = tips)


#sns.get_dataset_names()
```

**Box Plot:**

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

```python
#Boxplot

sns.boxplot(x ='day', y ='total_bill', data = tips, hue ='smoker')
```

**3.Distribution plots**

It is used for examining univariate and bivariate distributions.

We will be discussing 3 types of distribution plots namely:

distplot

joinplot

pairplot

```python
# Dis Plot - It is used basically for univariant set of observations and visualizes it through a histogram.

# set the background style of the plot
#sns.set_style('whitegrid')
#Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.
#Setting the hist flag to False in distplot will yield the kernel density estimation plot.

sns.distplot(tips['total_bill'], kde = True, color ='red', bins = 25)
```

```python
#Joinplot- Draw a plot of two variables with bivariate and univariate graphs.
sns.jointplot(x ='total_bill', y ='tip', data = tips)
```

```python
#Pairplot - It represents pairwise relation across the entire dataframe
#sns.pairplot(tips, hue ="sex")
sns.pairplot(tips.loc[:, ['total_bill','tip','sex','size']], hue="sex");
```

**4. Matrix plots**

A matrix plot is a color-coded diagram that has rows data, columns data, and values.

With the heatmap() function, we specify the matrix data that we want to plot.

```python
# Heat Maps

# correlation between the different parameters
tc = tips.corr()

# plot a heatmap of the correlated data
sns.heatmap(tc)
```

```python

sns.heatmap(tc, cmap=['red','green','blue'],vmin = .5 , vmax = 1.0,annot = True)
```
